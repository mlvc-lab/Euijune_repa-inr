```python
import torch
import torch.nn as nn
import numpy as np

import matplotlib.pyplot as plt

device = torch.device("cuda:7" if torch.cuda.is_available() else "cpu")
```

## The numer of encoder experiment

```python
ckpt = torch.load('logs/strainer/train/DIV2K/10img_mse/7_1_enc_depth_-1_seed_1234.pt')
print(ckpt.keys())
```

```python
for enc_num in [3, 5, 7, 9]:
    ckpt = torch.load(f'logs/strainer/eval/CelebA_HQ/from_10img_repair_depth_1/{enc_num}_1_enc_depth_-1_seed_1234.pt')
    ckpt_mse = torch.load(f'logs/strainer/eval/CelebA_HQ/from_10img_mse_depth_-1/{enc_num}_1_enc_depth_-1_seed_1234.pt')
    avg_psnrs = np.zeros(2000)
    for i in range(1, 21):
        i = str(i).zfill(2)
        avg_psnrs += ckpt[i]['psnr']
    avg_psnrs = [x / 20 for x in avg_psnrs]
    plt.plot(range(2000), avg_psnrs, label=f'encoder {enc_num} (REPA)')
    
    avg_psnrs = np.zeros(2000)
    for i in range(1, 21):
        i = str(i).zfill(2)
        avg_psnrs += ckpt_mse[i]['psnr']
    avg_psnrs = [x / 20 for x in avg_psnrs]
    plt.plot(range(2000), avg_psnrs, label=f'encoder {enc_num} (MSE)')

plt.legend()  
plt.show()
```

## Encoder depth
